## Step 2 - Setup and Library Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print("Libraries loaded!")
```

## Step 3 -  Data Loading and First Look

```python
# Loading the data 
df = pd.read_csv("/Users/divyammor/Desktop/wine+quality/A1_2528711_Mor-white.csv", sep=';')

df.head()
```

## Step 4 - Exploring the Data 

```python
# Understanding the dataset structure
df.info()
```

```python
# Mean
print("Mean:")
print(df.mean(numeric_only=True))
```

```python
# Median
print("Median:")
print(df.median(numeric_only=True))
```

```python
# Mode
print("Mode:")
print(df.mode().iloc[0])
```

## Step 5 -  Statistical Summary

```python
# Generating descriptive statistics for numerical features
df.describe()
```

## Step 6 -  Data Quality Assessment

```python
# Checking for missing values in the dataset
print("Missing values:")
print(df.isnull().sum())
```

```python
# Identifying and removing duplicate records from the dataset
print("Duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
# Analyzing the distribution of wine quality ratings
print("\nWine quality counts:")
print(df['quality'].value_counts().sort_index())
```

## Step 7 -  Correlation Analysis

```python
# Examining the correlation between features and wine quality
numeric_df = df.select_dtypes(include=[np.number])
print(numeric_df.corr()['quality'].sort_values(ascending=False))
```

## Step 8 - Visualization

```python
# Creating a histogram to analyze the frequency of wine quality ratings
fig, ax = plt.subplots(figsize=(8, 5))
df['quality'].hist(alpha=0.6, bins=7, color='gold', label='White wine', ax=ax)
ax.set_title('Distribution of Wine Quality Scores')
ax.set_xlabel('Quality Score')
ax.set_ylabel('Count')
ax.legend()
plt.tight_layout()
plt.savefig('fig1_quality_distribution.png', dpi=150)
plt.show()
```

```python
# Creating a boxplot to compare alcohol content across quality scores
fig, ax = plt.subplots(figsize=(9, 5))
df.boxplot(column='alcohol', by='quality', ax=ax)
ax.set_title('Alcohol Content by Quality Score')
ax.set_xlabel('Quality Score')
ax.set_ylabel('Alcohol (%)')
plt.suptitle('')
plt.tight_layout()
plt.savefig('fig2_alcohol_boxplot.png', dpi=150)
plt.show()
```

```python
# Creating a correlation heatmap to identify relationships between variables
fig, ax = plt.subplots(figsize=(10, 8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f',
            cmap='coolwarm', center=0, ax=ax, linewidths=0.5)
ax.set_title('Correlation Heatmap — Wine Quality Features')
plt.tight_layout()
plt.savefig('fig3_heatmap.png', dpi=150)
plt.show()
```
